```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
LE = LabelEncoder()
```

```python
LE = LabelEncoder()
for col in df.columns:
  if df[col].dtype == 'object':
    df[col] = LE.fit_transform(df[col])
```

```python
df.info()
```

```python
x  =df.drop('Churn', axis =1)
y = df['Churn']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train,y_test = train_test_split(x,y,test_size= 0.25)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier(n_estimators=50 ,max_depth= 3)
```

```python
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
#
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)*100
```

1ï¸âƒ£ n_estimators=50

ðŸ‘‰ Model 50 decision trees banayega.

Jitne zyada trees â†’ utna stable result

Bahut zyada trees â†’ slow training

Default usually 100 hota hai, yahan 50 use kiya gaya hai.

2ï¸âƒ£ max_depth=3

ðŸ‘‰ Har tree ki maximum depth 3 hogi.

Tree depth ka matlab:

Kitni levels tak decision lega

Depth kam hogi â†’ model simple hoga
Depth zyada hogi â†’ model complex hoga

Example:

Depth 3 â†’ kam overfitting

Depth 20 â†’ overfitting ka risk

```python
# max- depth , n_estimator ,....
param_grid = {
    'n_estimators': [100,200,300],
    'max_depth':  [1,3,5],
    'min_samples_split': [2,5,7],
    'min_samples_leaf': [1,2,4],
    'criterion': ['gini','entropy']
}
3*3*3*3*2
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
base_model = RandomForestClassifier(random_state=23)
```

```python
grid_search=GridSearchCV(estimator=base_model,param_grid=param_grid)
```

```python
grid_search.fit(x_train,y_train)
```

```python
final_model=RandomForestClassifier(n_estimators=100,max_depth=5,min_samples_leaf=1,min_samples_split=2,criterion='gini')
```

```python
final_model.fit(x_train,y_train)
```

```python
y_pred_final = final_model.predict(x_test)
```

```python
accuracy_score(y_pred_final,y_test)*100
```
